````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('previsao_de_renda.csv')
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
df.info()
````

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

#plot 1
freq = pd.crosstab(df['data_ref'], df['tipo_renda'])
freq = freq.div(freq.sum(axis = 1), axis = 0)
ax = freq.plot.bar(stacked = True, ax=axes[0,0])
ax.legend(bbox_to_anchor=(-0.05, 1), loc='upper right')

#plot 2
freq = pd.crosstab(df['data_ref'], df['educacao'])
freq = freq.div(freq.sum(axis = 1), axis = 0)
ax = freq.plot.bar(stacked = True, ax=axes[1,0])
ax.legend(bbox_to_anchor=(-0.05, 1), loc='upper right')

# plot 3
freq = pd.crosstab(df['data_ref'], df['estado_civil'])
freq = freq.div(freq.sum(axis = 1), axis = 0)
ax = freq.plot.bar(stacked = True, ax=axes[0,1])
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

# plot 4
freq = pd.crosstab(df['data_ref'], df['tipo_residencia'])
freq = freq.div(freq.sum(axis = 1), axis = 0)
ax = freq.plot.bar(stacked = True, ax=axes[1,1])
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

plt.subplots_adjust(hspace=0)
````

Foi percebido um aumento de 20% em casos de ensino superior completo

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

tick_labs = df['data_ref'].unique()

#plot 1
ax = sns.pointplot(x='data_ref', y='tempo_emprego', data=df, dodge=True, ci = 95, ax=axes[0,0])
labels = ax.set_xticklabels(tick_labs, rotation=90)

#plot 2
ax = sns.pointplot(x='data_ref', y='qt_pessoas_residencia', data=df, dodge=True, ci = 95, ax=axes[1,0])
labels = ax.set_xticklabels(tick_labs, rotation=45)

# plot 3
ax = sns.pointplot(x='data_ref', y='renda', data=df, dodge=True, ci = 95, ax=axes[0,1])
labels = ax.set_xticklabels(tick_labs, rotation=90)

# plot 4
ax = sns.pointplot(x='data_ref', y='idade', data=df, dodge=True, ci = 95, ax=axes[1,1])
labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.subplots_adjust(hspace=0)
````
